```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdate
import sys
sys.path.insert(1, '../src/')
import process_camels as pc

# Catchment 1
c1 = "camelsgb_33024.csv"
c2 = "camels_11124500.csv"
c3 = "hysets_10BE007.csv"

df1 = pc.process_input(c1)
df2 = pc.process_input(c2)
df3 = pc.process_input(c3)

# Create a figure that can hold multiple plots at once
# As an alternative you might want to use: https://seaborn.pydata.org/tutorial/axis_grids.html
fig, ax = plt.subplots(3, 1, figsize=(20, 12), sharex=True)


# Plot the runoff Q
sns.lineplot(data=df1, x="Date", y="Q [mm/day]", ax=ax[0], label="Q", color="orange")
sns.lineplot(data=df2, x="Date", y="Q [mm/day]", ax=ax[1], label="Q", legend=False, color="orange")
sns.lineplot(data=df3, x="Date", y="Q [mm/day]", ax=ax[2], label="Q", legend=False, color="orange")

# Plot the ET
s1 = sns.lineplot(data=df1, x="Date", y="PET [mm/day]", ax=ax[0], label="PET", color="green")
s2 = sns.lineplot(data=df2, x="Date", y="PET [mm/day]", ax=ax[1], label="PET", legend=False, color="green")
s3 = sns.lineplot(data=df3, x="Date", y="PET [mm/day]", ax=ax[2], label="PET", legend=False, color="green")

# Get the right hand side second y-axis
a1 = ax[0].twinx()
a2 = ax[1].twinx()
a3 = ax[2].twinx()
# Plot the precipitation as inverted bar chart
sns.barplot(data=df1, x="Date", y="P [mm/day]", ax=a1, label="P", color="#5a86ad")
sns.barplot(data=df2, x="Date", y="P [mm/day]", ax=a2, label="P", color="#5a86ad")
sns.barplot(data=df3, x="Date", y="P [mm/day]", ax=a3, label="P", color="#5a86ad")
a1.invert_yaxis()
a2.invert_yaxis()
a3.invert_yaxis()

# Show only the main ticks
locator = mdate.MonthLocator()
plt.gca().xaxis.set_major_locator(locator)

# display the figure to explore it
#plt.show()

# Rename the y-axis
s1.set(ylabel="[mm/day]")
s2.set(ylabel="[mm/day]")
s3.set(ylabel="[mm/day]")

## Additonal task: Add the temperature to the plot

s1.legend(bbox_to_anchor=(0, 1, 1, 0), loc="lower left")

# or save it directly
fig.savefig("catchment_stats.png")



```
