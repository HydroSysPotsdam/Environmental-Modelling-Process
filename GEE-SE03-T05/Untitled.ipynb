```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))

x = np.linspace(0, 30, 100)
Q = 3*sp.stats.gamma.pdf(x, 20)
plt.plot(x, 0.02 + Q, color="black")
plt.fill_between(x, 0, 0.2*Q + 0.02, color="gray", alpha=0.5)
plt.fill_between(x, 0.2*Q + 0.02, Q + 0.02, color="gray", alpha=0.25)
plt.text(19, 0.32,r"$Q = Q_\text{runoff} + Q_\text{base}$", fontsize=15, ha="right")
plt.text(19, 0.17,r"$Q_\text{runoff}$", fontsize=15, ha="center")
plt.text(19, 0.03,r"$Q_\text{base}$", fontsize=15, ha="center")
plt.text(25, 0.4,r"RR$ = \frac{Q_\text{runoff}}{P}$", fontsize=15)
plt.text(25.35, 0.35,r"BI$ = \frac{Q_\text{base}}{Q}$", fontsize=15)

plt.plot([9, 19], [0.45, 0.45], "-|", color="black")
plt.text(14, 0.47, "Lag Time", ha="center", fontsize=15)

x0 = x[80]
i0 = np.argmin(np.abs(x - x0))
s = np.gradient(Q, x)[i0]
plt.plot([x0 - 4, x0 + 4], [Q[i0] + 0.02 - s*4, Q[i0] + 0.02 + s*4], ls="--", color="C1")
plt.text(x0 - 0.5, Q[i0] + 0.05, "Recession Constant", color="C1", fontsize=15)

x = np.linspace(0, 30, 35)
P = 2*sp.stats.norm.pdf(x, 9, scale=2)
plt.bar(x, P, color="C0", zorder=99)
plt.text(4.5, 0.15, "Precipitation", color="C0", rotation=90, fontsize=15)

plt.xlim(2, 30)
plt.ylim(0, 0.5)
plt.axis("off")

plt.savefig("./figures/signatures.png", bbox_inches="tight", dpi=200)
```
