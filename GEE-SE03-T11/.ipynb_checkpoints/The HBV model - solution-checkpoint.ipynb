## 1 About HBV

The HBV (Hydrologiska Byråns Vattenbalansavdelning) model is a computer model simulating river runoff with minimum input data. It was developed to investigate river discharge and water pollution in Scandinavia. 
It is a [**spatially lumped**](https://en.wikipedia.org/wiki/Lumped-element_model) **bucket-type** catchment model that has relatively few model parameters and minimal forcing input requirements, using daily rainfall, temperature, and potential evaporation as input. For more information, check out [Seibert (1997)](https://iwaponline.com/hr/article/28/4-5/247/31640/Estimation-of-Parameter-Uncertainty-in-the-HBV) and the [Wikipedia page](https://en.wikipedia.org/wiki/HBV_hydrology_model).

The subroutines are: (1) snow routine (accumulation and melt), (2) soil moisture accounting and river routing.

<figure>
    <img src="./figures/HBV-fig.png" style="width:70%">
    <figcaption>Conceptual diagram of HBV (<a href="https://doi.org/10.3390/w12092376     ">Medina & Muñoz, 2020</a>)</figcaption>
</figure>

(1) The **snow routine** ...
* **expects** *time series* of *temperature* and *precipitation*, and the following *four parameters*:
|Parameter|Meaning|Units|Min|Max|
|---|---|---|---|---|
|Ts|threshold temperature for snow melt|°C|-3|3|
|CFMAX|degree day factor|mm/°Cd|0|20|
|CFR|refreezing factor |-|0|1|
|CWH|water holding capacity of snow|-|0|0.8|

* and **returns** these model results:
    * **Q_sim: time series of simulated river flow** (our focus in this course)
    * states: time series of simulated storages (soil water content ...)
    * fluxes: time series of simulated fluxes (actual evapotranspiration, recharge ...)

(2) The **soil moisture accounting and river routing** ...

* **expects** *time series* of *precipitation* and *potential evapotranspiration*, and the following *nine parameters*:
|Parameter|Meaning|Units|Min|Max|
|---|---|---|---|---|
|BETA|Exponential parameter in soil routine|-|0|7|
|LP|evapotranspiration limit|-|0.3|1|
|FC|field capacity|mm|1|2000|
|PERC|maximum flux from Upper to Lower Zone|mm/d|0|100|
|K0|near surface flow coefficient (ratio)|1/d|0.05|2|
|K1|upper Zone outflow coefficient (ratio)|1/d|0.01|1|
|K2|lower Zone outflow coefficient (ratio)|1/d|0.005|0.1|
|UZL|near surface flow threshold|mm|0|100|
|MAXBAS|flow routing coefficient|d|1|6|

    (Parameter bounds are adapted from [SAFE toolbox workflow](https://github.com/SAFEtoolbox/SAFE-python/blob/main/examples/workflows/workflow_eet_hbv.py))

* and **returns** these model results:
    * **Q_sim: time series of simulated river flow** (our focus in this course)
    * states: time series of simulated storages (soil water content ...)
    * fluxes: time series of simulated fluxes (actual evapotranspiration, recharge ...)

## 2 Using HBV

**Import packages**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdate
import numpy as np
import pandas as pd
import scipy.stats as st

import sys
sys.path.append("src/")
import HBV
from sampling import AAT_sampling # module to perform the input sampling
import plot_functions as pf # module to visualize the results
import RSA_thres as Rt # module to perform RSA with threshold


def calc_rmse(obs, sim):
    return np.sqrt(np.mean((obs - sim)**2))

def calc_abs_bias(obs, sim):
    return abs(np.mean(np.subtract(obs, sim)))

def hbv(par, precip, temp, evap):
    # Run HBV snow routine
    p_s, _, _ = HBV.snow_routine(par[:4], temp, precip)
    # Run HBV runoff simulation
    Case = 1 # for now we assume that the preferred path in the upper zone is runoff (Case = 1), it can be set to percolation (Case = 2)
    ini = np.array([0,0,0]) # initial state
    runoff_sim, _, _ = HBV.hbv_sim(par[4:], p_s, evap, Case, ini)
    return runoff_sim
    
def hbv_and_two_ofs(par, precip, temp, evap, runoff_obs, n_days, obj_fun1, obj_fun2):
    runoff_sim = hbv(par, precip, temp, evap)
    
    errors = np.nan * np.ones((2,))
    errors[0] = obj_fun1(runoff_obs[n_days:], runoff_sim[n_days:])
    errors[1] = obj_fun2(runoff_obs[n_days:], runoff_sim[n_days:])
    return errors, runoff_sim[n_days:]
```

**Read catchment data**

```python
# Read catchment data
catchment_names = ["Medina River, TX, USA", "Siletz River, OR, USA", "Trout River, BC, Canada"]
catchment_name = catchment_names[2]
# Read catchment data
file_dic = {catchment_names[0]: "camels_08178880", catchment_names[1]: "camels_14305500", catchment_names[2]: "hysets_10BE007"}
df_obs = pd.read_csv(f"data/{file_dic[catchment_name]}.csv")
# Make sure the date is interpreted as a datetime object -> makes temporal operations easier
df_obs.date = pd.to_datetime(df_obs['date'], format='%Y-%m-%d')
# Index frame by date
df_obs.set_index('date', inplace=True)
# Select only the columns we need
df_obs = df_obs[["total_precipitation_sum","potential_evaporation_sum","streamflow", "temperature_2m_mean"]]
# Rename variables
df_obs.columns = ["P [mm/day]", "PET [mm/day]", "Q [mm/day]", "T [C]"]
# Select time frame
start_date = '2002-10-01'
end_date = '2005-09-30'
df_obs = df_obs[start_date:end_date]
# Reformat the date for plotting
df_obs["Date"] = df_obs.index.map(lambda s: s.strftime('%b-%d-%y'))

# Prepare the time series intput
P = df_obs["P [mm/day]"].to_numpy()
PET = df_obs["PET [mm/day]"].to_numpy()
T = df_obs["T [C]"].to_numpy()
Q_obs = df_obs["Q [mm/day]"].to_numpy()
```

**Run HBV**

```python
n_spinup_days = 365
n_par_sets = 100
rmse_threshold = 3
bias_threshold = 1
dic_threshold = {'rmse': rmse_threshold, 'bias': bias_threshold}

str_par = ["Ts", "CFMAX", "CFR", "CWH", "BETA", "LP", "FC", "PERC", "K0", "K1", "K2", "UZL", "MAXBAS"]  # for parameter analysis

xmin = [-3, 0, 0, 0, 0, 0.3, 1, 0, 0.05, 0.01, 0.005, 0, 1]
xmax = [3, 20, 1, 0.8, 7, 1, 2000, 100, 2, 1, 0.1, 100, 6]

n_par = len(str_param)

distr_par = [np.nan] * n_par
for i in range(n_par):
    distr_par[i] = [xmin[i], xmax[i] - xmin[i]]

par_sets = AAT_sampling("lhs", n_par, st.uniform, distr_par, n_par_sets)

l_errors = []
par_set_min_rmse = par_sets[0]
min_rmse = 1000
for par_set in par_sets:
    # Run HBV snow simulation
    # hbv_func(par_set, precip, temp, evap, runoff_obs, n_days, obj_fun1, obj_fun2):
    (rmse, abs_bias), _ = hbv_and_two_ofs(par_set, P, T, PET, Q_obs, n_spinup_days, calc_rmse, calc_abs_bias)
    if rmse < min_rmse:
        min_rmse = rmse
        par_set_min_rmse = par_set
    l_errors.append((rmse,abs_bias))
    
np_errors = np.asarray(l_errors)
# RSA (find behavioural parameterizations, i.e. parameterizations with output values below the threshold):
mvd, spread, irr, idxb = Rt.RSA_indices_thres(par_sets, np_errors, [rmse_threshold, bias_threshold])

if idxb.sum() == 0:
    raise Exception("NO parameter combination resulted in behavioral model results (determined using objective function thresholds).")  
elif idxb.sum() == n_par_sets:
    raise Exception("ALL parameter combinations resulted in behavioral model results (determined using objective function thresholds).")  
else: 
    print(f"Behavorial parameter sets: {idxb.sum()}")
    print(f"Non-Behavorial parameter sets: {idxb.size - idxb.sum()}")
    i = 0
    for str_metric in dic_threshold.keys():
            plt.figure(figsize=(20,12))
            pf.scatter_plots(par_sets, np_errors[:, i], Y_Label=str_metric, X_Labels=str_par, idx=idxb)
            plt.show()
            i += 1

    plt.figure(figsize=(20, 12))
    Rt.RSA_plot_thres(par_sets, idxb, X_Labels=str_par, str_legend=['behav', 'non-behav'])
    plt.show()
    
    q_sim = hbv(par_set_min_rmse, P, T, PET)
    df_sim = pd.DataFrame({'Q_sim [mm/day]': q_sim, 'Date': df_obs["Date"].to_numpy()})
    
    plt.close()
    fig, ax = plt.subplots(1, 1, figsize=(20, 4))
    
    print("Plot the simulated Q with best rmse, and observed Q")
    sns.lineplot(data=df_sim[n_spinup_days:], x="Date", y="Q_sim [mm/day]", label="HBV")
    sns.lineplot(data=df_obs[n_spinup_days:], x="Date", y="Q [mm/day]", color="black", label="Observed")
    
    # Show only the main ticks
    locator = mdate.MonthLocator()
    plt.gca().xaxis.set_major_locator(locator)
    ax.set_title(catchment_name)
    plt.xticks(rotation=45)
    
    # Display the figure
    plt.show()
```
